```python
def drop_units(str_with_units):
    if isinstance(str_with_units, str):
        return float(str_with_units.split()[0])
    
car_data_dropped['mileage'] = car_data_dropped['mileage'].apply(drop_units)
car_data_dropped['engine'] = car_data_dropped['engine'].apply(drop_units)
car_data_dropped['max_power'] = car_data_dropped['max_power'].apply(drop_units)
```

```python
def get_torque(torque_str):
    if isinstance(torque_str, str):
        num = float(re.search(r'^[0-9]*\.?[0-9]+', torque_str)[0])
        if 'kgm' in torque_str:
            num *= 9.80665
        return num
```

```python
car_data_dropped['torque'] = car_data_dropped['torque'].apply(get_torque)
```

```python
def best_subset(y,param_list,df_data):
    full="+".join(param_list)
    output=[]
    length= len(param_list)
    full_model = smf.ols(f'{y}~{full}',data=df_data).fit()
    val=df_data[y]
    y_pred=full_model.fittedvalues
    p=length
    for i in range(length+1):
        for k in ["+".join(a) for a in combinations(param_list, i)]:
            if(len(k)!=0):
                #print('testing')
                reg = smf.ols(f'{y}~{k}',data=df_data).fit()
                #if(len(k.split('+'))<=p):
                #mal_cp=mallow.mallow(val,y_pred,reg.fittedvalues,p,len(k.split('+')))
                bp_test = het_breuschpagan(reg.resid, reg.model.exog)
               # print(k)
                output.append([k,reg.rsquared_adj,reg.aic,reg.bic,bp_test[1]])
                
    df_output=pd.DataFrame(output, columns=["Vars","ADJ R", "AIC", "BIC",'BP'])            
    return df_output
```

```python
y='selling_price_log'
param=['name','year','km_driven','fuel','seller_type','transmission','owner','mileage','max_power','torque','engine']
Dataset=best_subset(y,param,car_data_dropped)
```
